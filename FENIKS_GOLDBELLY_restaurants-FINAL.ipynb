```python
%pip install bs4
%pip install requests
from bs4 import BeautifulSoup
import requests

url = "https://www.goldbelly.com/featured-shops"
res = requests.get(url)
soup = BeautifulSoup(res.text)

restaurants = soup.select(".media-tile__location-link , .hidden-mobile , .media-tile__name")
print(restaurants)
```

```python
Data=[]
for item in soup.select(".media-tile__location-link"):
    all_txt=item.get_text(",", strip=True).split(",")
    if len(all_txt) != 2:
         Data.append(all_txt)
     
         
    
print(Data)
```

```python
#Including SOLD OUT
import pandas as pd
#Data=[]
name=[]
des=[]
ranking=[]
city=[]
state=[]
SOLD_OUT=[]
countn=0
countR=0

for item in soup.select(".media-tile__location-link"):
    all_txt=item.get_text(",", strip=True).split(",")
    if len(all_txt) != 2:
        if all_txt[0]!='SOLD OUT' and len(all_txt)==4:
            print(all_txt)
            name.append(all_txt[0])
            des.append(all_txt[1])
            city.append(all_txt[2])
            state.append(all_txt[3])
            ranking.append(None)
            SOLD_OUT.append(None)
            countn+=1
        elif all_txt[0]!='SOLD OUT' and len(all_txt)==6:
            name.append(all_txt[0])
            ranking.append(all_txt[2])
            des.append(all_txt[3])
            city.append(all_txt[4])
            state.append(all_txt[5])
            SOLD_OUT.append(None)
            countR+=1
        elif all_txt[0] == 'SOLD OUT':
            if len(all_txt)==7:
            #SOLD_OUT.append(all_txt) 
                name.append(all_txt[1])
                ranking.append(all_txt[3])
                des.append(all_txt[4])
                city.append(all_txt[5])
                state.append(all_txt[6])
                SOLD_OUT.append(all_txt[0])
            else:
                name.append(all_txt[1])
                ranking.append(None)
                des.append(all_txt[2])
                city.append(all_txt[3])
                state.append(all_txt[4])
                SOLD_OUT.append(all_txt[0])
                
                
                           
    
    

Data= pd.DataFrame({'name':name,'ranking':ranking,'description':des,'city':city,'state':state,'SOLD_OUT':SOLD_OUT})
print(Data)



```
